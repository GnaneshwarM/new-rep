```python
Prof_link =  []
url = "https://www.lsbu.ac.uk/study/study-at-lsbu/our-schools/engineering/subjects/computer-science-and-informatics/staff"
text = urllib.request.urlopen(url).read()
soup = BeautifulSoup(text)

data = soup.findAll('div',attrs={'class':'col-12 col-sm-8'})
for div in data:
    links = div.findAll('a',limit=1)
    for a in links:
        b = a['href']
        Prof_link.append(b)
```

```python
Prof_link =  []
url = "https://www.lsbu.ac.uk/study/study-at-lsbu/our-schools/engineering/subjects/computer-science-and-informatics/staff"
text = urllib.request.urlopen(url).read()
soup = BeautifulSoup(text)

data = soup.findAll('div',attrs={'class':'col-12 col-sm-8'})
for div in data:
    links = div.findAll('a', limit=1 )
    for a in links:
        b = a['href']
        Prof_link.append(b)
```

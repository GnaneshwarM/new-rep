```python
import requests 
import bs4
import lxml
from bs4 import BeautifulSoup
import pandas as pd
import re
import urllib.request
import schedule
import time
from datetime import datetime
```

```python
def primary():
    uni = requests.get('https://www.imperial.ac.uk/computing/research/data-science/')
    uni_imperial = 'https://www.imperial.ac.uk/computing/research/data-science/'
    uni_imperial2 = urllib.request.urlopen(uni_imperial)
    soup = BeautifulSoup(uni_imperial2)
    content = soup.find_all('span',attrs={'class':'person-name'})
    print("List of all the professors in the Imperial college as of",datetime.now())
    Prof =  []
    for i in content:
        profna = i.text
        bn = re.findall(r'\b[A-Z].*',profna)
        print(bn)
        bn = ''.join(bn)
        Prof.append(bn)
    print("Total number of professors:", len(content))
    print("Content Type:",type(content))
    key = soup.find_all('div',attrs={'class':'dept-wrapper'})
    print("The keywords required gathered for all the professors:")
    Prof_key =  []
    for i in key :
        profKa = i.text
        Kn = re.findall(r'\b[A-Z].*',profKa)
        print(Kn)
        Kn = ''.join(Kn)
        Prof_key.append(Kn)
    print("The Keywords length:",len(Prof_key))
    Prof_link =  []
    url = "https://www.imperial.ac.uk/computing/research/data-science/"
    text = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(text)
    data = soup.findAll('div',attrs={'class':'name-wrapper'})
    for div in data:
        links = div.findAll('a' ,attrs={'class':'name-link'} )
        for a in links:
            b = a['href']
            Prof_link.append(b)
    print(Prof_link)
    Prof_email =  []
    url = "https://www.imperial.ac.uk/computing/research/data-science/"
    text = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(text)
    data = soup.findAll('div',attrs={'class':'name-wrapper'})
    for div in data:
        links = div.findAll('a' ,attrs={'class':'email'} )
        for a in links:
            b = a['href']
            Prof_email.append(b)
    df = pd.DataFrame()
    import numpy as np
    uni = pd.DataFrame([np.nan] * 10)
    uni_name = uni.replace(np.nan, 'Imperial College London ', regex=True)
    print(uni_name)
    df['Name'] = Prof
    df['Expertise'] =Prof_key
    df['Email'] =  Prof_email
    df['Link'] = Prof_link
    df['University_Name']=uni_name
    print(df)
    df.to_csv(r'/Users/gnani/Documents/Candy 3 univ assignments/imperial.csv', index = True)

    
primary()
schedule.every(1).week.do(primary)
```
